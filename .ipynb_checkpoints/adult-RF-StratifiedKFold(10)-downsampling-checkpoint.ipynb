# StratifiedKFold instead of train/test set - downsampling

- it takes longer time. 
- better accuracy(for f-1 score)
- labelEncoding is reqired

```python
%%time
# data import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from IPython.display import display   

%matplotlib inline
URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

adult = pd.read_csv(
    URL,
    
    # specify the file encoding
    encoding="utf-8",
    
    # specify the separator in the data
    sep=",",      # comma separated values
    
    # ignore spaces after the separator
    skipinitialspace=True,
    index_col=None,
    
    # use manual headers
    header=None,
    names=[
        "age", "workclass", "fnlwgt", "education", 
        "education-num", "marital-status", "occupation",
        "relationship", "race", "sex", "capital-gain", 
        "capital-loss", "hours-per-week", "native-country",
        "wage"
    ]
)

# categorical -> one hot encoding
# convert each categorical feature using one-hot encoding

obj_df = adult.select_dtypes(include=["object"]).drop("wage", axis=1)
int_df = adult.select_dtypes(include=["int64"])

onehot_obj = pd.get_dummies(obj_df, drop_first=True)# dummy trap(one variable can be induced)
design_adult = pd.concat([onehot_obj,int_df], axis=1)

display(design_adult)

# labelEncoder 0,1
le = preprocessing.LabelEncoder()
le.fit(["<=50K", ">50K"])
```

```python
adult['wage'].value_counts(normalize=True).plot(kind='bar')
```

```python
adult['wage'].value_counts()
```

```python
design_adult['Class']=le.transform(adult['wage'])
```

```python
design_adult.head()
```

```python
# downsampling
from sklearn.utils import resample



df_majority = design_adult[design_adult.Class == 0]
df_minority = design_adult[design_adult.Class == 1]
```

```python
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=7841,     # to match minority class
                                 random_state=123) # reproducible results
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
# Display new class counts
df_downsampled.Class.value_counts()
```

```python

display(df_downsampled.reset_index().iloc[:,1:-1].head())

df_downsampled.reset_index().iloc[:,1:-1]

df_downsampled.reset_index().Class
```

```python
df_downsampled.iloc[:,:-1].head(1) #X variable all of the variables except Class variable
df_downsampled.Class
```

```python
# 속성(변수) 선택
X=df_downsampled.reset_index().iloc[:,1:-1]
y=df_downsampled.reset_index().Class
yhat = np.zeros(y.shape) # we will fill this with predictions
```

```python
# # 설명 변수 데이터를 정규화(normalization)
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)

from sklearn.model_selection import StratifiedKFold
# create cross validation iterator
cv = StratifiedKFold(n_splits=10)


from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(criterion='entropy', 
                             n_estimators=150,
                             min_samples_split=10,
                             min_samples_leaf=1,
                             max_features='auto',
                             oob_score=True,
                             random_state=10,
                             n_jobs=-1)


# NOTE: you can parallelize this using the cross_val_predict method
for train, test in cv.split(X,y):
    clf.fit(X[train],y[train])
    yhat[test] = clf.predict(X[test])
```

```python
    
from sklearn import metrics as mt    
total_accuracy = mt.accuracy_score(y, yhat)
print ('RF accuracy', total_accuracy)

# # train data 와 test data로 구분(7:3 비율)
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10) 
```

```python
print("%.4f" % clf.oob_score_)

# model 성능확인
from sklearn import metrics as mt


# 모형 성능 평가 - 평가지표 계산
report_rf = mt.classification_report(y, yhat)            
print(report_rf)
```

```python
# RF importances_
a=pd.concat((pd.DataFrame(design_adult.iloc[:,:-1].columns, columns = ['variable']), 
           pd.DataFrame(clf.feature_importances_, columns = ['importance'])), 
          axis = 1).sort_values(by='importance', ascending = False)[:20]
print('****Important vairables****','\n',a)
plt.barh(range(len(clf.feature_importances_)), clf.feature_importances_)
plt.show()

print ('Generalization score estimate from training data', clf.oob_score_)
```
